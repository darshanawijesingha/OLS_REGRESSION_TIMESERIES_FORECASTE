```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
from itertools import permutations

def evaluate_models(df, df_name):
    df = df.copy()
    results = {}

    columns = ['ordercount', 'sale', 'saleperorder']
    for target_col in columns:
        predictors = [col for col in columns if col != target_col]
        for predictor in predictors:
            X = df[[predictor]].values
            y = df[target_col].values

            # Try Linear, Poly2, Poly3
            models = {}

            # Linear
            lin_model = LinearRegression().fit(X, y)
            y_pred_lin = lin_model.predict(X)
            models['Linear'] = {
                'model': lin_model,
                'pred': y_pred_lin,
                'r2': r2_score(y, y_pred_lin),
                'mse': mean_squared_error(y, y_pred_lin),
                'formula': f"{target_col} = {lin_model.coef_[0]:.2f} * {predictor} + {lin_model.intercept_:.2f}"
            }

            # Polynomial Degree 2
            poly2 = PolynomialFeatures(degree=2)
            X_poly2 = poly2.fit_transform(X)
            model2 = LinearRegression().fit(X_poly2, y)
            y_pred2 = model2.predict(X_poly2)
            a, b = model2.coef_[2], model2.coef_[1]
            c = model2.intercept_
            models['Polynomial2'] = {
                'model': model2,
                'pred': y_pred2,
                'r2': r2_score(y, y_pred2),
                'mse': mean_squared_error(y, y_pred2),
                'formula': f"{target_col} = {a:.2f} * {predictor}^2 + {b:.2f} * {predictor} + {c:.2f}"
            }

            # Polynomial Degree 3
            poly3 = PolynomialFeatures(degree=3)
            X_poly3 = poly3.fit_transform(X)
            model3 = LinearRegression().fit(X_poly3, y)
            y_pred3 = model3.predict(X_poly3)
            a3, b3, c3 = model3.coef_[3], model3.coef_[2], model3.coef_[1]
            d3 = model3.intercept_
            models['Polynomial3'] = {
                'model': model3,
                'pred': y_pred3,
                'r2': r2_score(y, y_pred3),
                'mse': mean_squared_error(y, y_pred3),
                'formula': f"{target_col} = {a3:.4f} * {predictor}^3 + {b3:.4f} * {predictor}^2 + {c3:.4f} * {predictor} + {d3:.2f}"
            }

            # Find best model
            best_model_type = max(models, key=lambda k: models[k]['r2'])
            results[f"{predictor} ➜ {target_col}"] = {
                'input': predictor,
                'target': target_col,
                'model_type': best_model_type,
                'r2': models[best_model_type]['r2'],
                'mse': models[best_model_type]['mse'],
                'equation': models[best_model_type]['formula']
            }

    print(f"\n📊 Best Equations for {df_name}")
    for key, val in results.items():
        print(f"{key} ({val['model_type']}): {val['equation']} | R² = {val['r2']:.4f}")
    return results

# Run this on your cleaned dataframes
results_005 = evaluate_models(saledates_cut005, "Outlet 005")
results_009 = evaluate_models(saledates_cut009, "Outlet 009")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Step 1: Add 'day' index column
def add_day(df):
    df = df.copy()
    df['day'] = np.arange(len(df))
    return df

saledates_cut005 = add_day(saledates_cut005)
saledates_cut009 = add_day(saledates_cut009)
saledates025 = add_day(saledates025)

# Step 2: Combine training data
train_df = pd.concat([saledates_cut005, saledates_cut009], ignore_index=True)

# Step 3: Prepare training inputs
X_train = train_df[['ordercount', 'saleperorder']]
y_train = train_df['sale']
X_train_sm = sm.add_constant(X_train)  # add intercept

# Step 4: Fit OLS regression model
model = sm.OLS(y_train, X_train_sm).fit()

# Step 5: Print model summary (for p-values, R², etc.)
print(model.summary())

# Step 6: Predict sale for saledates025
X_test = saledates025[['ordercount', 'saleperorder']]
X_test_sm = sm.add_constant(X_test)
saledates025['predicted_sale'] = model.predict(X_test_sm)

# Step 7: Find first day with predicted_sale >= 1M
threshold_day = saledates025[saledates025['predicted_sale'] >= 1_000_000]

if not threshold_day.empty:
    hit_day = threshold_day['day'].iloc[0]
    print(f"\n✅ Outlet 025 is predicted to hit 1M daily sales on day {hit_day} "
          f"(index: {hit_day}, date approx: Day {hit_day+1})")
else:
    print("\n❌ Outlet 025 is NOT predicted to hit 1M daily sales in the existing 49-day data")

# Step 8: Plot predicted vs actual sales
plt.figure(figsize=(10, 5))
plt.plot(saledates025['day'], saledates025['sale'], label='Actual Sale (025)', color='green')
plt.plot(saledates025['day'], saledates025['predicted_sale'], label='Predicted Sale', color='blue', linestyle='--')
plt.axhline(1_000_000, color='red', linestyle='--', label='1M Threshold')
if not threshold_day.empty:
    plt.axvline(hit_day, color='purple', linestyle='--', label=f'Predicted 1M Day: {hit_day}')
plt.title("Daily Sale Forecast for Outlet 025")
plt.xlabel("Day")
plt.ylabel("Sale (LKR)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Step 1: Add 'day' index
def add_day(df):
    df = df.copy()
    df['day'] = np.arange(len(df))
    return df

# Add day column
saledates_cut005 = add_day(saledates_cut005)
saledates_cut009 = add_day(saledates_cut009)
saledates025 = add_day(saledates025)

# Step 2: Combine training data (cut005 + cut009)
train_df = pd.concat([saledates_cut005, saledates_cut009], ignore_index=True)

# Step 3: Prepare input for regression model
X_train = train_df[['ordercount', 'saleperorder']]
y_train = train_df['sale']
X_train_sm = sm.add_constant(X_train)

# Step 4: Fit the OLS regression model
model = sm.OLS(y_train, X_train_sm).fit()

# Step 5: Print summary
print(model.summary())

# Step 6: Start with existing 025 data
df_forecast = saledates025.copy()

# Step 7: Forecast until predicted daily sale >= 1M
day = df_forecast['day'].max() + 1
while True:
    # Get last known values to estimate next
    last = df_forecast.iloc[-1]
    new_ordercount = last['ordercount'] * 1.01  # simulate 1% growth
    new_saleperorder = last['saleperorder'] * 1.005  # simulate slight growth

    # Prepare input
    new_input = pd.DataFrame({
        'ordercount': [new_ordercount],
        'saleperorder': [new_saleperorder]
    })
    new_input_sm = sm.add_constant(new_input, has_constant='add')
    new_sale = model.predict(new_input_sm)[0]

    # Append forecasted day
    df_forecast = pd.concat([df_forecast, pd.DataFrame([{
        'day': day,
        'ordercount': new_ordercount,
        'saleperorder': new_saleperorder,
        'sale': np.nan,
        'predicted_sale': new_sale
    }])], ignore_index=True)

    # Stop when we reach 1M sale
    if new_sale >= 1_000_000:
        print(f"\n✅ Predicted day to reach 1M daily sales: Day {day} "
              f"(~ in {day - saledates025['day'].max()} future days)")
        break

    day += 1

# Step 8: Plot actual + predicted sales
plt.figure(figsize=(10, 6))
plt.plot(df_forecast['day'], df_forecast['sale'], label='Actual Sale (025)', color='green', marker='o')
plt.plot(df_forecast['day'], df_forecast['predicted_sale'], label='Predicted Sale', color='blue', linestyle='--')
plt.axhline(1_000_000, color='red', linestyle='--', label='1M Threshold')
plt.axvline(day, color='purple', linestyle='--', label=f'1M Sale Day: {day}')
plt.title("Forecasting Daily Sale for Outlet 025 Until 1M Reached")
plt.xlabel("Day")
plt.ylabel("Daily Sale (LKR)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
